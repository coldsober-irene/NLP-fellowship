```python
import requests
from bs4 import BeautifulSoup as bs

pageLink = "https://www.geeksforgeeks.org/how-to-scrape-paragraphs-using-python/?ref=lbp&id=discuss"
articleLink = "https://igihe.com/ikoranabuhanga/article/u-rwanda-rwageze-ku-ikoranabuhanga-rya-google-street-view"

page = requests.get(pageLink)
soup = bs(page.text, "html.parser")
soup.prettify
body = soup.body
# div corps
corp = body.find_all("ul", class_="leftBarList")
```

```python
parent_div = body.find("div", class_ = "sideBar--wrap")
print(parent_div)
```

```python
# GET THE LINKS IN 
links = parent_div.find_all("a")
print(links)
```

```python
# get text from links
for link in links:
  print(link.get_text())
```

```python
from bs4 import BeautifulSoup as bs
import requests

def get_data(url):
  data = requests.get(url)
  return data.text

def parse_html(htmlData):
  parsed = bs(htmlData, 'html.parser')
  return parsed

htmlData = get_data("https://www.geeksforgeeks.org/how-to-scrape-paragraphs-using-python/")
parsed = parse_html(htmlData)
# print(parsed.prettify())
# h1 = parsed.find_all('h1')
# print(h1[0].get_text())
# GET MAIN DIV
b = parsed.body
main_div = b.find_all("div")
for div in main_div:
  for i in div:
    print(i)
    print()

```
